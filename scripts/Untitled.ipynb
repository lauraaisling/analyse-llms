```python
from transformers import GPTNeoXForCausalLM, AutoTokenizer
import torch
# torch.cuda.is_available()
```

```python
device = "cuda:0" 
```

```python
model = GPTNeoXForCausalLM.from_pretrained("lomahony/eleuther-pythia70m-hh-sft",).to(device)

tokenizer = AutoTokenizer.from_pretrained("lomahony/eleuther-pythia70m-hh-sft",)#.to(device)

inputs = tokenizer("Hello, I am", return_tensors="pt").to(device)
tokens = model.generate(**inputs)
tokenizer.decode(tokens[0])
```

```python
text = "Hello, I am"
input_ids = tokenizer.encode_plus(text)["input_ids"]
input_ids
```

```python
input_tokens = torch.tensor(input_ids).long().to(device)
input_tokens
```

```python
model(torch.unsqueeze(input_tokens,0), return_dict=True)
```
